```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rc_file("./matplotlibrc")

pgf_with_latex = {
    "pgf.texsystem": "pdflatex",
    "pgf.rcfonts"  : False,
}

other_settings = {
    'xtick.minor.visible': True,
    'xtick.direction' : 'in',
    'ytick.minor.visible': True,
    'ytick.direction' : 'in',
    'xtick.labelsize': 26.0,
    'xtick.labelsize': 26.0,
    'ytick.labelsize': 26.0,
    'axes.labelsize': 26.0,
    'axes.titlesize': 26.0,
    'axes.xmargin': 0.05,
    'axes.ymargin': 0.05,
    'figure.figsize': [8,6]
}
matplotlib.rcParams.update(pgf_with_latex)
matplotlib.rcParams.update(other_settings)
```

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.plot(x,y)
plt.xticks()

fig = plt.gcf()
print(fig.get_size_inches())
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.savefig("plots/short_tic_labels.pgf")
plt.savefig("plots/short_tic_labels.pdf")
plt.show()
```
